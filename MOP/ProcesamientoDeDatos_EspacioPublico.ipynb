```python
import pandas as pd
import geopandas as gpd
from functools import reduce
```

## Datos Gobiernos Locales

```python
df_goblocales = gpd.read_file("../GobiernosLocales/GobiernosLocales.shp")
df_goblocales.rename(columns={'Provincia':'provincia', 'Departamen':'departamen', 'Municipio':'municipio', 'UTA2020':'uta2020'}, inplace=True)

df_goblocales = df_goblocales.to_crs("EPSG:3035")
df_goblocales['area_total_ha'] = df_goblocales.area / 1000000
df_goblocales = df_goblocales.to_crs("EPSG:4326")
```

```python
# carga de datos de las dimensiones Inserción Territorial y Sistema Demográfico
df_goglocal_pob =  gpd.read_file("../CapasGeograficasOriginales/InsercionTerritorial_GobiernosLocales.shp")
df_goglocal_sistdem = pd.read_csv("../BD_Originales/SistemaDemografico_GobiernosLocales.csv")
df_goglocal_sistdem.rename(columns={'coduta':'uta2020'}, inplace=True)

df_goglocal_sistdem.uta2020 = df_goglocal_sistdem.uta2020.astype(str)
for i in range(len(df_goglocal_sistdem)):
    if len(df_goglocal_sistdem.loc[i, 'uta2020']) == 8:
        df_goglocal_sistdem.loc[i, 'uta2020'] = '0'+df_goglocal_sistdem.loc[i, 'uta2020']
```

## Espacios Verdes

```python
# carga de espacios verdes
df_espaciosverdes = gpd.read_file("../CapasGeograficasOriginales/EspaciosVerdes.geojson") 
```

```python
# cruce de los espacios verdes con los gobiernos locales 
df_inter_gobespverdes = gpd.overlay(df_goblocales, df_espaciosverdes, how='intersection')
# cálculo del área
df_inter_gobespverdes = df_inter_gobespverdes.to_crs("EPSG:3035")
df_inter_gobespverdes['sup_verde'] = df_inter_gobespverdes.area
df_inter_gobespverdes = df_inter_gobespverdes.to_crs("EPSG:4326")
# exrpotación del resultado
df_inter_gobespverdes.to_file("../CapasGeograficasCruzadas/espaciosVerdes_goblocal.shp")
```

```python
# lipieza del capo gna
df_inter_gobespverdes['gna_mod'] = df_inter_gobespverdes['gna']
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Plaza Principal', 'gna_mod'] = 'Plaza'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Municipal', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Urbano', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Recreativo', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Público', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Plazoletas', 'gna_mod'] = 'Plazoleta'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Regional', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Nacional', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Paseo Recreativo', 'gna_mod'] = 'Paseo'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Provincial', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Natural', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque de la Ciudad', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Jardines', 'gna_mod'] = 'Jardín'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parquecito', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parques', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Vivero Municipal', 'gna_mod'] = 'Vivero'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Natural Municipal', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Plaza Central', 'gna_mod'] = 'Plaza'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Jardín Botánico', 'gna_mod'] = 'Jardín'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Paseo Municipal', 'gna_mod'] = 'Paseo'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Plaza Estación', 'gna_mod'] = 'Plaza'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Paque', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Faunístico', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Infantil', 'gna_mod'] = 'Parque'
df_inter_gobespverdes.loc[df_inter_gobespverdes.gna_mod == 'Parque Recreativo y Municipal', 'gna_mod'] = 'Parque'

df_inter_gobespverdes.loc[(df_inter_gobespverdes.gna_mod != 'Plaza') & (df_inter_gobespverdes.gna_mod != 'Plazoleta') 
    &(df_inter_gobespverdes.gna_mod != 'Parque') &(df_inter_gobespverdes.gna_mod != 'Paseo'), 'gna_mod'] = 'Otro'
```

```python
# Generación de los valores de las distintas variable
# superficie de espacios verdes
df_sup_gobespverdes = pd.DataFrame(df_inter_gobespverdes.groupby("uta2020")["sup_verde"].sum()).reset_index()
# población de espacios verdes
df_goblocales_espverde_pob = df_sup_gobespverdes.merge(df_goglocal_pob[['uta2020', 'k2_sup_tot']], how='left', on="uta2020")
df_goblocales_espverde_pob = df_goblocales_espverde_pob.merge(df_goglocal_sistdem[['uta2020', 'dem_pobt']], how='left', on="uta2020")
df_goblocales_espverde_pob.rename(columns={'k2_sup_tot':'area_ha','dem_pobt':'pob_gob'}, inplace=True) 
df_goblocales_espverde_pob['pob_gob'] = df_goblocales_espverde_pob['pob_gob'].fillna(0)
df_goblocales_espverde_pob['sup_espverdexpob'] = 0
df_goblocales_espverde_pob.loc[df_goblocales_espverde_pob.pob_gob > 0, 'sup_espverdexpob'] = df_goblocales_espverde_pob.loc[df_goblocales_espverde_pob.pob_gob > 0,'sup_verde'] / df_goblocales_espverde_pob.loc[df_goblocales_espverde_pob.pob_gob > 0,'pob_gob']
# superficie de espacios verdes por habitante
df_goblocales_espverde_pob['espverde_est'] = "Insuficiente"
df_goblocales_espverde_pob.loc[df_goblocales_espverde_pob.sup_espverdexpob > 10, 'espverde_est'] = "Buena"
df_goblocales_espverde_pob.loc[df_goblocales_espverde_pob.sup_espverdexpob > 15, 'espverde_est'] = "Óptima"
df_goblocales_espverde_pob.drop(columns=['area_ha','pob_gob'], inplace=True)
# tipo de espacios verdes
df_goblocales_espverde_tipo = df_inter_gobespverdes.pivot_table(index="uta2020", columns="gna_mod", aggfunc='size').reset_index().fillna(0)
```

```python
# generación del dataframe de espacios verdes
dfs_verde= [df_goblocales_espverde_pob, df_goblocales_espverde_tipo]
dfs_verde_merged = reduce(lambda  left,right: pd.merge(left,right,on=['uta2020'], how='outer'), dfs_verde)
# cambio del nombre de las columnas
dfs_verde_merged.rename(columns={'sup_verde':'ev_sup', 'sup_espverdexpob':'ev_supxhab', 'espverde_est':'ev_sup_cat', 'Otro':'tip_otro',
    'Parque':'tip_parque', 'Paseo':'tip_paseo', 'Plaza':'tipo_plaza', 'Plazoleta':'tip_plazol'}, inplace=True)
# extracción de los nombres de los campos
df_columns_evp = list(dfs_verde_merged.columns)

# cruce de los datos generados con los gobienros locales
df_goblocales_espaciosverdes = df_goblocales.merge(dfs_verde_merged, on='uta2020', how='left')
df_goblocales_espaciosverdes.ev_sup.fillna(0, inplace=True)
df_goblocales_espaciosverdes.ev_supxhab.fillna(0, inplace=True)
df_goblocales_espaciosverdes.tip_otro.fillna(0, inplace=True)
df_goblocales_espaciosverdes.tip_parque.fillna(0, inplace=True)
df_goblocales_espaciosverdes.tip_paseo.fillna(0, inplace=True)
df_goblocales_espaciosverdes.tipo_plaza.fillna(0, inplace=True)
df_goblocales_espaciosverdes.tip_plazol.fillna(0, inplace=True)
# df_goblocales_espaciosverdes.ev_sup_por.fillna(0, inplace=True)
# formateo de los campos numéricos
df_goblocales_espaciosverdes.ev_sup = round(df_goblocales_espaciosverdes.ev_sup, 2)
df_goblocales_espaciosverdes.ev_supxhab = round(df_goblocales_espaciosverdes.ev_supxhab, 1)
df_goblocales_espaciosverdes.tip_otro = df_goblocales_espaciosverdes.tip_otro.astype(int)
df_goblocales_espaciosverdes.tip_parque = df_goblocales_espaciosverdes.tip_parque.astype(int)
df_goblocales_espaciosverdes.tip_paseo = df_goblocales_espaciosverdes.tip_paseo.astype(int)
df_goblocales_espaciosverdes.tipo_plaza = df_goblocales_espaciosverdes.tipo_plaza.astype(int)
df_goblocales_espaciosverdes.tip_plazol = df_goblocales_espaciosverdes.tip_plazol.astype(int)
# df_goblocales_espaciosverdes.ev_sup_por = df_goblocales_espaciosverdes.ev_sup_por * 100
# df_goblocales_espaciosverdes.ev_sup_por = round(df_goblocales_espaciosverdes.ev_sup_por, 2)

# eliminación de columnas y seteo de los datos de los gob locales sin polígono a "null"
df_goblocales_espaciosverdes.loc[df_goblocales_espaciosverdes['buffer'] == 1, 'ev_sup_cat'] = None
df_goblocales_espaciosverdes.loc[df_goblocales_espaciosverdes['buffer'] == 1, ['ev_sup', 'ev_supxhab', 'tip_otro', 'tip_parque', 'tip_paseo', 'tipo_plaza', 'tip_plazol']] = float("NaN")
```

```python
# exportación de los datos generados a csv y shp
df_goblocales_espaciosverdes[df_columns_evp].to_csv("../BD_Procesada/EspacioPublico_GobiernosLocales.csv", encoding='utf-8')
schema_evp_dic={'geometry': 'Polygon', 'properties': {'provincia':'str', 'departamen':'str', 'municipio':'str', 'tipgobloc':'str', 'uta2020':'str',
                    'ev_sup':'float:10.1', 'ev_supxhab':'float:10.1', 'ev_sup_cat':'str', 'tip_otro':'int', 'tip_parque':'int', 'tip_paseo':'int', 'tipo_plaza':'int', 'tip_plazol':'int'}}
df_goblocales_espaciosverdes[['provincia', 'departamen', 'municipio', 'tipgobloc']+ df_columns_evp+['geometry']].to_file(
    "../CapasGeograficasProcesadas/EspacioPublico_GobiernosLocales.shp", 
    driver='ESRI Shapefile', 
    schema=schema_evp_dic,
    encoding='utf-8')
df_goblocales_espaciosverdes.head()
```
